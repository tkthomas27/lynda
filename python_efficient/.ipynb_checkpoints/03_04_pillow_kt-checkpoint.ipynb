## Manipulating images with Pillow

```python
download = requests.get(imgurl)
```

```python
imgright = PIL.Image.open(io.BytesIO(download.content))
```

```python
imgright
```

```python
imgright_resized = imgright.resize((400,400))
```

```python
imgright
```

```python
imgright.dewarped = imgright_resized.transform((400,300),PIL.Image.QUAD,data = (0,0,100,400,300,400,400,0), resample = PIL.Image.BILINEAR)
```
